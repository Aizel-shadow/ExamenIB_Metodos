```python
def secant_method(f, x0, x1, tol=1e-6, max_iter=100):    
    x_prev = x0
    x_curr = x1
    iter_count = 0

    while abs(f(x_curr)) > tol and iter_count < max_iter:
        # Calculate the next approximation using the secant method formula
        x_next = x_curr - f(x_curr) * (x_curr - x_prev) / (f(x_curr) - f(x_prev))

        # Update variables for the next iteration
        x_prev = x_curr
        x_curr = x_next
        iter_count += 1

    return x_curr, iter_count
```

```python
import math

def secant_method(f, x0, x1, tol=1e-6, max_iter=100):
    x_prev = x0
    x_curr = x1
    f_prev = f(x_prev)  
    f_curr = f(x_curr)  
    iter_count = 0
    call_count = 2  
    
    while abs(f_curr) > tol and iter_count < max_iter:
       
        x_next = x_curr - f_curr * (x_curr - x_prev) / (f_curr - f_prev)
        
      
        x_prev = x_curr
        f_prev = f_curr  
        
        x_curr = x_next
        f_curr = f(x_curr)  
        call_count += 1
        
        iter_count += 1
    
    return x_curr, iter_count, call_count


def f1(x):
    return x**3 + 2*x**2 + 3*x - 1


def f2(x):
    return 35 * math.sin(x * math.pi) + 10


```

Pruebas 

```python
x0, x1 = 2, 3

result1 = secant_method(f1, x0, x1)
print(f" x³ + 2x² + 3x - 1 = 0")
print(f"  Raíz: {result1[0]:.10f}")
print(f"  Iteraciones: {result1[1]}")
print(f"  Llamadas a la función: {result1[2]}")
```

```python
result2 = secant_method(f2, x0, x1)
print(f"35·sin(x·π) + 10 = 0")
print(f"  Raíz: {result2[0]:.10f}")
print(f"  Iteraciones: {result2[1]}")
print(f"  Llamadas a la función: {result2[2]}")
```
